## 1. Why Calculus is Needed in Machine Learning

Machine Learning models learn by:
- Making predictions
- Measuring error (loss)
- Updating parameters to **reduce error**

The key question:
> *In which direction should we change parameters to reduce loss?*

Derivatives answer this question.



## 2. What is a Derivative? (Intuition)

A **derivative** measures:
> How fast a quantity changes when input changes

Geometric meaning:
- Slope of a curve at a point

ML meaning:
- How sensitive the loss is to model parameters


````python

import numpy as np
import matplotlib.pyplot as plt
````


## 3. Derivative as Slope

Consider a simple function:
```
f(x) = xÂ²
```

Its slope:
- Negative for x < 0
- Zero at x = 0
- Positive for x > 0


````python

x = np.linspace(-5, 5, 200)
y = x**2

plt.plot(x, y)
plt.axhline(0)
plt.axvline(0)
plt.title("f(x) = xÂ²")
plt.show()
````


## 4. Numerical Derivative (No Formulas)

We can estimate slope using:
```
(f(x+h) âˆ’ f(x)) / h
```

This is how computers *think* about derivatives.


````python

def numerical_derivative(f, x, h=1e-5):
    return (f(x+h) - f(x)) / h

f = lambda x: x**2

for point in [-2, -1, 0, 1, 2]:
    print(point, numerical_derivative(f, point))
````


## 5. Analytical Derivative (Just the Result)

For:
```
f(x) = xÂ²
```
The derivative is:
```
f'(x) = 2x
```

We care about **interpretation**, not memorization.



## 6. Loss Function Example (ML Core)

Consider a simple loss:
```
L(w) = (w âˆ’ 3)Â²
```

Goal:
Find `w` that minimizes the loss.


````python

w = np.linspace(-2, 6, 200)
loss = (w - 3)**2

plt.plot(w, loss)
plt.title("Loss Function L(w)")
plt.xlabel("w")
plt.ylabel("Loss")
plt.show()
````


## 7. Derivative Tells Direction of Change

- If derivative > 0 â†’ decrease `w`
- If derivative < 0 â†’ increase `w`
- If derivative = 0 â†’ minimum or maximum

This is the foundation of **gradient descent**.


````python

dL_dw = 2*(w - 3)
plt.plot(w, dL_dw)
plt.axhline(0, color='red')
plt.title("Derivative of Loss")
plt.show()
````


## 8. ML Interpretation (Very Important)

Derivative answers:
> *How should the parameter change to reduce error?*

Without derivatives:
- No learning
- No optimization
- No ML
