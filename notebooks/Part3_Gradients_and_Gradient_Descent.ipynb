## 1. Why Gradients Matter in Machine Learning

Real ML models have:
- Many parameters
- One loss function

We need a **single direction** that tells us how to change **all parameters**
to reduce the loss.

That direction is called the **gradient**.


```python

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Gradient = Vector of Partial Derivatives

For a loss function:
L(w‚ÇÅ, w‚ÇÇ, ..., w‚Çô)

The gradient is:
‚àáL = [‚àÇL/‚àÇw‚ÇÅ, ‚àÇL/‚àÇw‚ÇÇ, ..., ‚àÇL/‚àÇw‚Çô]

The gradient always points in the **direction of steepest increase**.



## 3. Loss Surface Intuition

Consider the loss:
L(w‚ÇÅ, w‚ÇÇ) = (w‚ÇÅ ‚àí 2)¬≤ + (w‚ÇÇ ‚àí 1)¬≤

This creates a **bowl-shaped surface**.


```python

def loss(w1, w2):
    return (w1 - 2)**2 + (w2 - 1)**2

w1 = np.linspace(-1, 5, 100)
w2 = np.linspace(-1, 5, 100)
W1, W2 = np.meshgrid(w1, w2)
Z = loss(W1, W2)

plt.contour(W1, W2, Z, levels=30)
plt.xlabel("w1")
plt.ylabel("w2")
plt.title("Loss Contours")
plt.show()
```


## 4. Computing the Gradient

For the loss:
L(w‚ÇÅ, w‚ÇÇ) = (w‚ÇÅ ‚àí 2)¬≤ + (w‚ÇÇ ‚àí 1)¬≤

Gradient:
‚àáL = [2(w‚ÇÅ ‚àí 2), 2(w‚ÇÇ ‚àí 1)]


```python

def gradient(w1, w2):
    return np.array([2*(w1 - 2), 2*(w2 - 1)])

print("Gradient at (0,0):", gradient(0,0))
```


## 5. Why Gradient Descent Works

- Gradient points **uphill**
- We want to go **downhill**
- So we move in the **negative gradient direction**



## 6. Gradient Descent Update Rule

Update rule:
w = w ‚àí Œ± ¬∑ ‚àáL

Where:
- Œ± = learning rate


```python

# Gradient Descent implementation
w = np.array([0.0, 0.0])
alpha = 0.1

trajectory = []

for _ in range(30):
    trajectory.append(w.copy())
    w = w - alpha * gradient(w[0], w[1])

trajectory = np.array(trajectory)

plt.contour(W1, W2, Z, levels=30)
plt.plot(trajectory[:,0], trajectory[:,1], marker='o')
plt.title("Gradient Descent Path")
plt.xlabel("w1")
plt.ylabel("w2")
plt.show()

print("Final parameters:", w)
```


## 7. Learning Rate (Œ±)

- Too small ‚Üí slow learning
- Too large ‚Üí overshooting / divergence

Choosing Œ± is critical in ML.



## 8. ML Interpretation (Very Important)

Gradient descent:
- Minimizes loss
- Learns parameters
- Powers linear regression, logistic regression, neural networks

This is the **engine of Machine Learning**.
