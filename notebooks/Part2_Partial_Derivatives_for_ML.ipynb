## 1. Why Partial Derivatives Are Needed

In real ML models:
- Loss depends on **many parameters**
- We must know how loss changes with **each parameter separately**

Partial derivatives answer:
> *How does the loss change if only one parameter changes?*


```

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Multi-Variable Loss Function

Example:
L(w‚ÇÅ, w‚ÇÇ) = (w‚ÇÅ ‚àí 2)¬≤ + (w‚ÇÇ ‚àí 1)¬≤


```

def loss(w1, w2):
    return (w1 - 2)**2 + (w2 - 1)**2
```


## 3. Loss Surface Visualization


```

w1 = np.linspace(-1, 5, 100)
w2 = np.linspace(-1, 5, 100)
W1, W2 = np.meshgrid(w1, w2)
Z = loss(W1, W2)

plt.contour(W1, W2, Z, levels=30)
plt.xlabel("w1")
plt.ylabel("w2")
plt.title("Loss Contours")
plt.show()
```


## 4. Partial Derivatives

‚àÇL/‚àÇw‚ÇÅ = 2(w‚ÇÅ ‚àí 2)  
‚àÇL/‚àÇw‚ÇÇ = 2(w‚ÇÇ ‚àí 1)


```

def dL_dw1(w1, w2):
    return 2 * (w1 - 2)

def dL_dw2(w1, w2):
    return 2 * (w2 - 1)

print(dL_dw1(0,0), dL_dw2(0,0))
```


## 5. Numerical Partial Derivatives


```

def numerical_partial(f, w1, w2, var="w1", h=1e-5):
    if var == "w1":
        return (f(w1+h, w2) - f(w1, w2)) / h
    else:
        return (f(w1, w2+h) - f(w1, w2)) / h

print(numerical_partial(loss, 0, 0, "w1"))
print(numerical_partial(loss, 0, 0, "w2"))
```


## 6. ML Interpretation

Partial derivatives form the **gradient**, guiding optimization.
